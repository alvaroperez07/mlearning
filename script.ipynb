Importing dataset and taking a look

```python
import pandas as pd
#Preview Dataset
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.describe())
```

```python
#Preview datacolumns to select variables
print (data.columns)
```

```python
#Example selecting one column

melbourne_price_data = data.SalePrice
melbourne_price_data.head()
```

```python
#Example selecting two columns
select_columns = [ 'SalePrice', 'BedroomAbvGr']
two_columns_data = data[select_columns]
two_columns_data.describe()
```

```python
#Selecting the SalePrice as the variable we want to predict
y = data.SalePrice
```

```python
#Selecting variables to build the model 
melbourne_predict = ['PoolArea', 'MiscVal', 'MoSold', 'YrSold','Id', 'MSSubClass',  'LotArea' ,  'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd',  '1stFlrSF', '2ndFlrSF',       'LowQualFinSF', 'GrLivArea', 'FullBath',       'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr','TotRmsAbvGrd','Fireplaces','WoodDeckSF', 'OpenPorchSF',       'EnclosedPorch', '3SsnPorch', 'ScreenPorch']
x = data[melbourne_predict]
x.describe()
```

```python
from sklearn.model_selection import train_test_split

#Create training set y validation set
train_x, val_x, train_y, val_y = train_test_split(x,y,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor

#Define tree model 
melbourne_model = DecisionTreeRegressor()

#Fit tree Model
melbourne_model.fit(train_x,train_y)
```

```python
#Using the model to predict some houses
print ("Making predictions for the following 5 houses")
print(val_x.head())
print("The predictions are")
print(melbourne_model.predict(val_x.head()))
```

```python
#Defining min_mae for iterations
min_mae=1000000
```

```python
#Fit tree Model with leaves
melbourne_model = DecisionTreeRegressor (max_leaf_nodes = 50, random_state = 0)
melbourne_model.fit(train_x,train_y)
```

```python
from sklearn.metrics import mean_absolute_error

#Validate model, save 
predicted_home_prices = melbourne_model.predict(val_x)
mae = mean_absolute_error(val_y,predicted_home_prices)
print ("The mae for the model is: " , mae, "and last mae was: ",min_mae, ". The difference is: ",mae-min_mae)
if mae<min_mae:
    min_mae = mae
    model_min = melbourne_model
    print ("Model saved")
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor (max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return (mae)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_x,val_x, train_y, val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

#Define ran
forest_model = RandomForestRegressor()
forest_model.fit(train_x, train_y)
melb_preds = forest_model.predict (val_x)
print (mean_absolute_error(val_y, melb_preds))
```

```python
mae = mean_absolute_error(val_y,melb_preds)
print ("The mae for the model is: " , mae, "and last mae was: ",min_mae, ". The difference is: ",mae-min_mae)
if mae<min_mae:
    min_mae = mae
    model_min = melbourne_model
    print ("Model saved")
```
